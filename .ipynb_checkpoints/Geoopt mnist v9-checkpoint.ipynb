```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

import matplotlib.pyplot as plt
from torchvision.transforms.functional import rotate

import numpy as np
from torch.func import grad
from scipy.linalg import lstsq, null_space

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
def G(d): # questa è  rot_mat
    mat = np.zeros(d**4).reshape(d**2,d**2)
    for j in range(d):
        for i in range(d):
            k = (i+1)*d-j
            h = i+(j)*d+1

            mat[h-1][k-1] = 1
    return mat

def W(output_dim, d = 4):
    kernel = null_space((np.identity(d*d) -  G(d)).T).T
    W = []
    for j in range(output_dim):
        vec = np.zeros(d**2)
        for k in kernel:
            vec = (np.random.random()-0.5)*2*k + vec # kernel[j]
        W.append(list(vec/np.linalg.norm(vec)))
    W = np.array(W)
    return np.array(W)

def projector(d = 4):
    kernel = null_space((np.identity(d*d) -  G(d)).T)
    
    return torch.Tensor(kernel @ kernel.T)
```

```python
# Impostazioni principali
batch_size = 64
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset e DataLoader
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_dataset = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root="./data", train=False, transform=transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Estrai un'immagine dal dataset
image, label = train_dataset[0]  # Prendi la prima immagine
image = image.squeeze(0)         # Rimuovi la dimensione del canale (1, 28, 28 -> 28, 28)

# Ruota l'immagine usando PIL
image_pil = transforms.ToPILImage()(image)  # Converte il tensor in immagine PIL
rotated_image_pil = image_pil.rotate(90)    # Ruota l'immagine di 45 gradi

# Plot dell'immagine originale e ruotata
plt.figure(figsize=(8, 4))

# Immagine originale
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title(f"Originale - Label: {label}")
plt.axis("off")

# Immagine ruotata
plt.subplot(1, 2, 2)
plt.imshow(rotated_image_pil, cmap="gray")
plt.title("Ruotata di 90°")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Modello: Rete Neurale a 3 layer con G immutabile
class ThreeLayerNNWithMatrix(nn.Module):
    def __init__(self, caste_matrix):
        super(ThreeLayerNNWithMatrix, self).__init__()
        self.register_buffer("caste_matrix", caste_matrix)  # Matrice G fissa (non aggiornata)
        self.fc1 = nn.Linear(28 * 28, 128, bias=True)         # Primo layer: da 784 a 128 neuroni
        self.fc2 = nn.Linear(128, 64)                        # Secondo layer: da 128 a 64 neuroni
        self.fc3 = nn.Linear(64, 10)                         # Terzo layer: da 64 a 10 neuroni (classi MNIST)
        self.relu = nn.LeakyReLU()                              # Funzione di attivazione

    def forward(self, x):
        x = x.view(x.size(0), -1)                           # Appiattisce l'immagine (batch_size, 28*28)
        
        # Modifica W del primo layer con la moltiplicazione G @ W
        modified_weight = torch.matmul(self.caste_matrix, self.fc1.weight.T)
        x = torch.matmul(x, modified_weight) + self.fc1.bias  # Calcolo manuale del primo layer
        x = self.relu(x)                                     # Attivazione ReLU
        
        x = self.relu(self.fc2(x))                          # Secondo layer + ReLU
        x = self.fc3(x)                                     # Terzo layer (output)
        return x
```

```python
# Creazione della matrice personalizzata G
input_size = 28 * 28  # Dimensione dell'input (784)
custom_matrix = projector(d = 28).to(device)  # Matrice identità di esempio (immutabile)
learning_rate = 0.01
num_epochs = 60

# Inizializza il modello, la loss e l'ottimizzatore
model = ThreeLayerNNWithMatrix(custom_matrix).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
losses = []
accs = []
accs_val = []
invariances = []

# Ciclo di addestramento
print("Inizio addestramento...")
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for images, labels in train_loader:
        
        images, labels = images.to(device), labels.to(device)
        images_90 = transforms.functional.rotate(images, 90)
        images_180 = transforms.functional.rotate(images, 180)
        images_270 = transforms.functional.rotate(images, 270)

        outputs = model(images)
        outputs_90 = model(images_90)
        outputs_180 = model(images_180)
        outputs_270 = model(images_270)

        _, predicted = torch.max(outputs, 1)
        _, predicted_90 = torch.max(outputs_90, 1)
        _, predicted_180 = torch.max(outputs_180, 1)
        _, predicted_270 = torch.max(outputs_270, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        correct_1 = (predicted == predicted_90).sum().item()
        correct_2 = (predicted == predicted_180).sum().item()
        correct_3 = (predicted == predicted_270).sum().item()
        
        
        inv = (correct_1+ correct_2+ correct_3)/(3*len(labels))
        
        loss = criterion(outputs, labels)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        
    acc = correct/total
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}")
    print(f"Accuracy: {acc} ")
    print(f"Invariance: {inv} \n")
    
    losses.append(running_loss/len(train_loader))
    accs.append(acc)
    invariances.append(inv)
          
    if epoch == 45:
        optimizer = optim.SGD(model.parameters(), lr=learning_rate/10)

    if epoch == 55:
        optimizer = optim.SGD(model.parameters(), lr=learning_rate/100)
        
    # Valutazione sul test set
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accs_val.append(correct / total)

    print(f"Precisione sul test set: {100 * correct / total:.2f}% \n")
```

```python
# Crea una figura con due subplot affiancati
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 riga, 2 colonne, dimensione figura

# Primo plot
axes[0].plot(np.array(losses), label = "Loss")
axes[0].set_title("Loss")
axes[0].set_xlabel("epoch")
axes[0].set_ylabel("loss")

# Secondo plot
axes[1].scatter(range(len(accs)),accs, label = "Accuracy train")
axes[1].scatter(range(len(accs_val)),accs_val, label = "Accuracy test")
axes[1].set_title("Accuracy")
axes[1].set_xlabel("epoch")
axes[1].set_ylabel("Accuracy")

# Mostra i plot
plt.tight_layout()  # Per evitare sovrapposizioni
axes[1].legend()
axes[0].legend()
plt.show()
```

```python
# Supponiamo di avere un DataLoader chiamato test_loader
images_list = []
labels_list = []

# Itera attraverso il DataLoader
for images, labels in train_loader:
    images_list.append(images)  # Accumula le immagini
    labels_list.append(labels)  # Accumula le etichette

# Concatena tutti i batch per creare un unico tensore
all_images = torch.cat(images_list, dim=0).to(device)  # Dimensione: [numero_totale_immagini, canali, altezza, larghezza]
all_labels = torch.cat(labels_list, dim=0).to(device)  # Dimensione: [numero_totale_immagini]

print(f"Tensore immagini: {all_images.shape}")
print(f"Tensore etichette: {all_labels.shape}")
```

```python
images_90 = torch.stack([transforms.functional.rotate(image, 90) for image in all_images[0:]]).to(device)
images_180 = torch.stack([transforms.functional.rotate(image, 180) for image in all_images[0:]]).to(device)
images_270 = torch.stack([transforms.functional.rotate(image, 270) for image in all_images[0:]]).to(device)

# Passaggio attraverso il modello
outputs = model(all_images[0:])
outputs_90 = model(images_90)
outputs_180 = model(images_180)
outputs_270 = model(images_270)

_, predicted = torch.max(outputs, 1)
_, predicted_90 = torch.max(outputs_90, 1)
_, predicted_180 = torch.max(outputs_180, 1)
_, predicted_270 = torch.max(outputs_270, 1)

total = all_labels[0:].size(0)
correct = (predicted == all_labels[0:]).sum().item()

correct = (predicted == all_labels[0:]).sum().item()
correct_1 = (predicted == predicted_90).sum().item()
correct_2 = (predicted == predicted_180).sum().item()
correct_3 = (predicted == predicted_270).sum().item()

invariance = (correct_1+ correct_2+ correct_3)/(3*len(all_labels))
print("Invariance", invariance)
print("Accuracy",correct/total)
```

```python
cm = confusion_matrix(all_labels.cpu(), predicted.cpu(), labels=range(10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=range(10))
disp.plot()
plt.show()
```

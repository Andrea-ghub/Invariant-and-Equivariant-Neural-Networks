```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funzione per calcolare accuratezza
def calculate_accuracy(loader, model, rotate=None):
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            # Ruota se specificato
            if rotate:
                images = torch.stack([transforms.functional.rotate(image, rotate) for image in images])
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return 100 * correct / total

# Funzione per confrontare predizioni su immagini normali e ruotate
def compare_predictions(loader, model, rotations):
    with torch.no_grad():
        for rotate in rotations:
            same_predictions = 0
            total_images = 0
            for images, _ in loader:
                images = images.to(device)
                # Predizioni su immagini normali
                outputs_normal = model(images)
                _, preds_normal = torch.max(outputs_normal, 1)

                # Predizioni su immagini ruotate
                images_rotated = torch.stack([transforms.functional.rotate(image, rotate) for image in images])
                outputs_rotated = model(images_rotated)
                _, preds_rotated = torch.max(outputs_rotated, 1)

                # Confronta predizioni
                same_predictions += (preds_normal == preds_rotated).sum().item()
                total_images += images.size(0)

            print(f"Rotazione di {rotate}°: {same_predictions}/{total_images} predizioni uguali "
                  f"({100 * same_predictions / total_images:.2f}%)")
```

```python
# Configurazioni
batch_size = 64
num_epochs = 5
learning_rate = 0.001
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Trasformazioni per il training (con rotazione casuale di 90°, 180°, 270°)
train_transform = transforms.Compose([
    transforms.RandomChoice([
        transforms.RandomRotation(degrees=(90, 90)),
        transforms.RandomRotation(degrees=(180, 180)),
        transforms.RandomRotation(degrees=(270, 270))
    ]),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Trasformazioni per il test (senza rotazioni)
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Dataset e DataLoader
train_dataset = datasets.MNIST(root="./data", train=True, transform=train_transform, download=True)
test_dataset = datasets.MNIST(root="./data", train=False, transform=test_transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Modello
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Appiattisce l'immagine
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Inizializzazione modello
model = SimpleNN().to(device)

# Ottimizzatore e funzione di perdita
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# Training
for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Aggiorna le metriche
        train_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_accuracy = 100 * correct / total
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {train_loss:.4f}, Accuracy: {train_accuracy:.2f}%")
```

```python
# Valutazione finale
model.eval()

# Calcola accuratezza su dataset normale
normal_accuracy = calculate_accuracy(test_loader, model, rotate=None)
print(f"\nAccuracy su dataset normale: {normal_accuracy:.2f}%")

# Calcola accuratezza su dataset ruotato di 90°, 180°, 270°
rotated_90_accuracy = calculate_accuracy(test_loader, model, rotate=90)
rotated_180_accuracy = calculate_accuracy(test_loader, model, rotate=180)
rotated_270_accuracy = calculate_accuracy(test_loader, model, rotate=270)

print(f"Accuracy su dataset ruotato di 90°: {rotated_90_accuracy:.2f}%")
print(f"Accuracy su dataset ruotato di 180°: {rotated_180_accuracy:.2f}%")
print(f"Accuracy su dataset ruotato di 270°: {rotated_270_accuracy:.2f}%")

# Confronta predizioni su immagini normali e ruotate
print("\nConfronto tra predizioni su immagini normali e ruotate:")
compare_predictions(test_loader, model, rotations=[90, 180, 270])
```

```python
# Funzione per calcolare le predizioni e le etichette
def get_predictions_and_labels(loader, model, rotate=None):
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            # Ruota se specificato
            if rotate:
                images = torch.stack([transforms.functional.rotate(image, rotate) for image in images])
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    return np.array(all_preds), np.array(all_labels)
```

```python
# Funzione per plottare la confusion matrix
def plot_confusion_matrix(preds, labels, title):
    cm = confusion_matrix(labels, preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```

```python
# Dataset normale
print("\nDataset normale:")
normal_preds, normal_labels = get_predictions_and_labels(test_loader, model, rotate=None)
normal_accuracy = (normal_preds == normal_labels).mean() * 100
print(f"Accuracy: {normal_accuracy:.2f}%")
plot_confusion_matrix(normal_preds, normal_labels, "Confusion Matrix - Dataset Normale")
```

```python
# Dataset ruotato di 90°
print("\nDataset ruotato di 90°:")
rotated_90_preds, rotated_90_labels = get_predictions_and_labels(test_loader, model, rotate=90)
rotated_90_accuracy = (rotated_90_preds == rotated_90_labels).mean() * 100
print(f"Accuracy: {rotated_90_accuracy:.2f}%")
plot_confusion_matrix(rotated_90_preds, rotated_90_labels, "Confusion Matrix - Rotazione 90°")
```

```python
# Dataset ruotato di 180°
print("\nDataset ruotato di 180°:")
rotated_180_preds, rotated_180_labels = get_predictions_and_labels(test_loader, model, rotate=180)
rotated_180_accuracy = (rotated_180_preds == rotated_180_labels).mean() * 100
print(f"Accuracy: {rotated_180_accuracy:.2f}%")
plot_confusion_matrix(rotated_180_preds, rotated_180_labels, "Confusion Matrix - Rotazione 180°")
```

```python
# Dataset ruotato di 270°
print("\nDataset ruotato di 270°:")
rotated_270_preds, rotated_270_labels = get_predictions_and_labels(test_loader, model, rotate=270)
rotated_270_accuracy = (rotated_270_preds == rotated_270_labels).mean() * 100
print(f"Accuracy: {rotated_270_accuracy:.2f}%")
plot_confusion_matrix(rotated_270_preds, rotated_270_labels, "Confusion Matrix - Rotazione 270°")
```
